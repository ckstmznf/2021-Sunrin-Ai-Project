- 주제 선정 배경 : 최근 스타벅스에 관한 글을 하나 보았는데 스타벅스에 대한 부정적인 글이었다. 스타벅스를 자주 가지 않는 성격이라 스타벅스가 어떠니 잘 모르겠어서 사람들의 유형별로 스타벅스에 대한 만족도가 어떻게 되는지 알아보기 위해서 선정하였다.
- 데이터 분석 과정 설명 : 우선 어떤 분야의 사람이 가장 많이 오는지를 알아보았다. <br>가장 많이 오는 분야의 사람들을 기반으로 만족도를 알아보았다.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

plt.rc("font", family="Malgun Gothic")
plt.rc('font', size = 12)
plt.style.use(["seaborn-deep"])
plt.figure(figsize=(15, 6), dpi=100)
```

```python
df = pd.read_csv("Starbucks satisfactory survey encode cleaned_10413박희찬_10422조상진.csv")
df.head()
```

## 3. 탐색적 데이터 분석 과정

- 과정 순서대로 코드와 내용 설명하세요.


#### 3-1. 데이터 전처리
- 전처리 내용 기재, 엑셀파일 내에서 처리했을 경우에도 내용 기재

> 데이터에 NaN 값이 있는 행은 없다.

>  기존에 gender 열이 0과 1로 남녀를 구분지었다면 남자를 mael, 여자를 female로 변경

> 기존에 나이가 0 ~ 3까지 숫자로 나이의 범위를 지정했다면 실질적인 나이의 범위를 표시

> 기존에 방문주기의 숫자가 클수록 조금 오는 것이었다면 작을수록 조금 오도록 변경

> 소비자의 각각의 만족도를 보여주는 열인 17 ~ 23열의 평균값을 rateMeen 열에 추가했다.

```python
pd.DataFrame(df.isna().sum())
```

```python
df["gender"] = df["gender"].replace(0, "male")
df["gender"] = df["gender"].replace(1, "female")


df["age"] = df["age"].replace(0, "10대 이하")
df["age"] = df["age"].replace(1, "20대")
df["age"] = df["age"].replace(2, "30대")
df["age"] = df["age"].replace(3, "40대 이상")


df["status"] = df["status"].replace(0, "학생")
df["status"] = df["status"].replace(1, "자영업자")
df["status"] = df["status"].replace(2, "취업")
df["status"] = df["status"].replace(3, "주부")

df["income"] = df["income"].replace(0, "~ 7,100,000원")
df["income"] = df["income"].replace(1, " 7,100,000원 ~ 14,200,000원")
df["income"] = df["income"].replace(2, "14,200,000원 ~ 28,400,000원")
df["income"] = df["income"].replace(3, "28.400,000원 ~ 42,600,000원")
df["income"] = df["income"].replace(4, "42,600,000원 ~")

df["method"] = df["method"].replace(0, "매장")
df["method"] = df["method"].replace(1, "드라이브스루")
df["method"] = df["method"].replace(2, "테이크 아웃")
df["method"] = df["method"].replace(3, "Never")
df["method"] = df["method"].replace(4, "기타")

df["visitNo"] = 3 - df["visitNo"]

df.head()
```

```python
df["rateMeen"] = df.iloc[:, 17:24].mean(axis = 1)

df.head()
```

#### 3-2. 표를 활용하여 데이터 분석하기
- 분석별 결과 내용이 항상 있어야 함.
- 2개 이상의 표를 활용하여 분석하기

```python
dfGedner = pd.DataFrame(df.groupby(by = "gender").count()["Id"])
dfGedner = dfGedner.sort_values(by="Id", ascending=False)
dfGedner

# 여성이 59명, 남성이 54명으로 여성이 더 많이 온다.
```

```python
dfAge = pd.DataFrame(df.groupby(by="age").count()["Id"])
dfAge = dfAge.sort_values(by="Id", ascending=False)
dfAge

# 20대가 79명으로 가장 많이 오고 그 다음으로 30대, 10대, 40대 순이다.
```

```python
dfStatus = pd.DataFrame(df.groupby(by="status").count()["Id"])
dfStatus = dfStatus.sort_values(by="Id", ascending=False)
dfStatus

# 직장에 다니고 있는 사람이 가장 많이 오고 학생, 자영업자, 주부 순이다.
```

```python
dfIncome = pd.DataFrame(df.groupby(by="income").count()["Id"])
dfIncome = dfIncome.sort_values(by="Id", ascending=False)
dfIncome

# 1년 소득이 7100,000원인 사람이 가장 많았다.
# 그 이유는 사회 초년생인 20대가 많이 있기때문이라고 생각 된다.
```

```python
genderMaxPeople = df["gender"] == dfGedner.index[0]
ageMaxPeople = df["age"] == dfAge.index[0]
statusMaxPeople = df["status"] == dfStatus.index[0]
incomeMaxPeople = df["income"] == dfIncome.index[0]

maxPeople = df[genderMaxPeople & ageMaxPeople & statusMaxPeople & incomeMaxPeople]

print(f"가장 많이 오는 집단에 속한 사람(여성이면서 20대, 직장인, 연 소득 7,100,000원 이하)은\n{len(maxPeople)}명입니다.\n")
for i in maxPeople["Id"]:
    print(f"{i}번", end = " ")
```

```python
maxPeople[["visitNo"]]

# 가장 많이 오는 집단의 사람의 방문 주기
# 값이 클수록 많이 온다.
```

#### 3-3.그래프를 활용하여 분석하기
- 분석별 결과 내용이 항상 있어야 함
- 서로 다른 그래프 3개와 결과 내용 (그래프 작성시 제목, x, y축 레이블 달기)

```python
plt.figure(figsize=(15, 6), dpi=100)
plt.title("고객 성별 비율")

colors = []
if(dfGedner.index[0] == "male"):
    colors.append("skyblue")
    colors.append("pink")
else:
    colors.append("pink")
    colors.append("skyblue")

plt.pie(
    dfGedner["Id"],
    labels=dfGedner.index,
    autopct="%.1f%%",
    colors=colors,
)
plt.legend()

plt.show()

# 고객 비율이 여성이 52.2%로 남성보다 4.4% 더 많이 온다.
```

```python
plt.figure(figsize=(15, 6), dpi=100)

plt.title("고객 연령 비율")

plt.pie(
    dfAge["Id"],
    labels= dfAge.index,
    autopct="%.1f%%",
    colors=["salmon", "lime", "pink", "aqua"],
    explode = [0.05, 0.05, 0.05, 0.05],
    counterclock=False
)

plt.legend(loc="lower right")

plt.show()

# 고객중 20대가 69%로 가장 많이 오고 20대가 과반수 이상이다.
```

```python
plt.figure(figsize=(15, 6), dpi=100)
plt.title("고객 직업 수")

plt.hist(df["status"], color="salmon")

plt.show()

# 스타벅스 고객중에서는 취업한 직장인이 가장 많이 온다.
```

```python
plt.figure(figsize=(15, 6), dpi=100)
plt.title("고객의 연 소득")
plt.rc('font', size = 15)

plt.hist(df["income"], color="orange")
plt.xlabel("소득 범위")
plt.ylabel("고객 수")

plt.show()

# 1년 소득이 7100,000원인 사람이 가장 많았다.
```

```python
# 방문 방법에 따른 분류
methodInShop = df[df["method"] == "매장"]
methodTakeOut = df[df["method"] == "테이크 아웃"]

plt.rc('font', size = 30)
plt.figure(figsize=(25, 12), dpi=100)

plt.title("방문 방법 별 만족도")

plt.hist(
    methodInShop["rateMeen"],
    bins = 15,
    label = "매장 방문",
    alpha = 0.7
)

plt.hist(
    methodTakeOut["rateMeen"], 
    bins = 15,
    label = "테이크 아웃",
    alpha = 0.7
)

plt.xlabel("만족도")
plt.ylabel("고객수")
plt.legend()

plt.show()


# 방문한 방법 별로 각 만족도를 선택한 고객수의 히스토그램
# 어떻게 방문했는지에 따라 만족도 차이는 없어 보인다.
```

```python
plt.figure(figsize=(7, 4), dpi=100)

plt.rc('font', size = 12)
plt.title("방문 주기 별 만족도")


dfSometime = df[df["visitNo"] == 0]
dfMonth = df[df["visitNo"] == 1]
dfWeeks = df[df["visitNo"] == 2]
dfAllDay = df[df["visitNo"] == 3]

# dfSometime["rateMeen"]

plt.bar(
    ["가끔 방문", "매달 방문", "매주 방문", "매일 방문"],
    [dfSometime["rateMeen"].mean(), dfMonth["rateMeen"].mean(), dfWeeks["rateMeen"].mean(), dfAllDay["rateMeen"].mean()],
    color = ["lightcoral", "wheat", "plum", "peachpuff"]
)

plt.xlabel("메장에 방문하는 정도")
plt.ylabel("만족도")

plt.show()

# 방문주기에 따른 평균 만족도
# 가끔 방문하는 사람들의 평균 만족도는 3.5점이다.
```

```python
plt.figure(figsize=(10, 5), dpi=100)
plt.title("지출 금액과 만족도")

dfSpend0 = df[df["spendPurchase"] == 0]
dfSpend1 = df[df["spendPurchase"] == 1]
dfSpend2 = df[df["spendPurchase"] == 2]
dfSpend3 = df[df["spendPurchase"] == 3]


plt.bar(
    ["0원", " ~ 5000원", "5000원 ~ 10000원", "10000원 이상"],
    [dfSpend0["rateMeen"].mean(), dfSpend1["rateMeen"].mean(), dfSpend2["rateMeen"].mean(), dfSpend3["rateMeen"].mean()],
    color = ["lightcoral", "wheat", "plum", "peachpuff"]
)

plt.xlabel("지출 정도 (값이 클수록 많이 지출)")
plt.ylabel("만족도 평균")

plt.show()

# 지출한 정도에 따른 만족도
# 0원을 지출한 사람의 만족도는 3.5점이다.
```

```python
students = df.copy()[df["age"] == "20대"]

plt.figure(figsize=(10, 6), dpi=100)

plt.title("20대 소비 보기")

plt.hist(students["spendPurchase"], color="skyblue")

plt.xticks([0, 1, 2, 3])
plt.xlabel('0원 | ~ 5000원 | 5000원 ~ 10000원 | 10000원 이상')
plt.ylabel("사람 수")

plt.show()

# 20대인 고객들이 어느정도를 소비했는지 알아보았다.
# 5000원을 지출한 사람이 가장 많았다.
```

## 4. 결론
- 질문에 대한 답
- 알게된 점, 느낀점 등

## 결론
- 가장 많이 오는 부류
- - 여성
- - 20대
- - 회사원
- - 연 소득이 7,100,000원 이하인 사람

> 특히 연 소득이 7,100,000원 미하인 사람이 많은 이유는 회사원도 많이 오지만 학생들도 많이 오기때문이라고 예상된다.

- 매장 방문과 테이크아웃 하는 사람의 만족도 차이는 없는것같다.
- 방문 주기와 만족도의 상관도는 데이터의 수가 적어서 분석하기가 어렵다.<br>
하지만 방문 주기가 많을수록 만족도도 올라간다고 유추 가능
- 지출 금액과 만족도의 상관도 역시 데이터의 수가 적어서 분석하기 어렵다.
- 학생의 경우 6000원 이하로 소비하는 사람이 대부분이고 6000원 ~ 12000원 


## 느낀점
데이터의 수가 더 많거나, 소비자 조사의 대상 국가가 한국이었으면 더 좋겠다고 생각한다.<br>
데이터 수가 적어서 여기서 나온 결과가 정답이라고는 할 수 없을것 같다.
